```python
data1 = df.pivot_table('Country',index='Continent',aggfunc='count')
data1
```

# Châu Âu nhiều quốc gia nhất

```python
data2 = df.pivot_table('GDP (millions of US$)',index='Continent',aggfunc='sum')
data2
```

```python
data3=pd.merge(data1,data2, on='Continent')
data3
```

```python
data3['AVG GDP per Continent'] = data3['GDP (millions of US$)']/data3['Country']
data3
```
